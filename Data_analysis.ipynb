```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
p_covid= "/content/covid19_Confirmed_dataset.csv"
corona_csv = pd.read_csv(p_covid)
corona_csv.head()
```

```python
corona_csv.shape
```

```python
corona_csv.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
corona_csv.head(10)
```

```python
corona_data_agg = corona_csv.groupby("Country/Region").sum()
```

```python
corona_data_agg.head()
```

```python
corona_data_agg.shape
```

```python
corona_data_agg.loc["China"][1:].plot(label='China')
corona_data_agg.loc["India"][1:].plot(label='India')

plt.legend()
plt.title('COVID-19 Cases in China and India')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)

plt.show()
```

```python
corona_data_agg.loc['China'][1:].plot()
```

```python
corona_data_agg.loc['India'][1:].plot()
```

```python
corona_data_agg = corona_data_agg.apply(pd.to_numeric, errors='coerce')

corona_data_agg.ffill(inplace=True)
corona_data_agg.loc['China'].diff().plot()


plt.xlabel('Date')
plt.ylabel('Difference')
plt.title('Daily Change in COVID-19 Cases for China')

plt.grid(True)
plt.show()
```

```python
china_data = corona_data_agg.loc['China'].diff().max()
```

```python
india_data = corona_data_agg.loc['India'].diff().max()
```

```python
pakistan_data = corona_data_agg.loc['Pakistan'].diff().max()
```

```python
countries = list(corona_data_agg.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_data_agg.loc[c].diff().max())
corona_data_agg["max_infection_rate"] = max_infection_rates
```

```python
corona_data_agg.head()
```

```python
corona_data = pd.DataFrame(corona_data_agg["max_infection_rate"])
```

```python
corona_data.head()
```

```python
from matplotlib import pyplot as plt
corona_data['max_infection_rate'].plot(kind='line', figsize=(8, 4), title='max_infection_rate')
plt.gca().spines[['top', 'right']].set_visible(False)
```
